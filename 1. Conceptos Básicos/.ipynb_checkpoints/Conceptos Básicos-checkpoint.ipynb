```python
import cv2
from IPython.display import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
def obtenerImagenRGBCanalizada(imagen, canal):
    imagenCanalizada = imagen.copy()
    if canal == "R":
        imagenCanalizada[:, :, 1] = 0
        imagenCanalizada[:, :, 2] = 0
    elif canal == "G":
        imagenCanalizada[:, :, 0] = 0
        imagenCanalizada[:, :, 2] = 0
    elif canal == "B":
        imagenCanalizada[:, :, 0] = 0
        imagenCanalizada[:, :, 1] = 0
    return imagenCanalizada

def construirListasCoordenadasDeMatriz(matriz, numeroFilas, numeroColumnas):
    listaCoordenadasX = []
    listaCoordenadasY = []
    listaCoordenadasZ = []
    for filaActual in range(numeroFilas):
        for columnaActual in range(numeroColumnas):
            listaCoordenadasX.append(filaActual)
            listaCoordenadasY.append(columnaActual)
            listaCoordenadasZ.append(matriz[filaActual, columnaActual])
    return listaCoordenadasX, listaCoordenadasY, listaCoordenadasZ
```

```python
rutaImagen = 'Figures/Flores.jpg'
anchoVentanaCentral = 29
altoVentanaCentral = 23

imagenColor = cv2.imread(rutaImagen, -1)
imagenColor = cv2.cvtColor(imagenColor, cv2.COLOR_BGR2RGB)
imagenGrises = cv2.imread(rutaImagen, -1)
imagenGrises = cv2.cvtColor(imagenGrises, cv2.COLOR_BGR2GRAY)

altoImagen, anchoImagen = imagenColor.shape[:2]
print("La imagen {} tiene tama√±o {}x{}".format(rutaImagen, anchoImagen, altoImagen))

imagenColorZonaCentral = imagenColor[int((altoImagen - altoVentanaCentral) / 2) : int((altoImagen + altoVentanaCentral) / 2), int((anchoImagen - anchoVentanaCentral) / 2) : int((anchoImagen + anchoVentanaCentral) / 2)]
imagenGrisesZonaCentral = imagenGrises[int((altoImagen - altoVentanaCentral) / 2) : int((altoImagen + altoVentanaCentral) / 2), int((anchoImagen - anchoVentanaCentral) / 2) : int((anchoImagen + anchoVentanaCentral) / 2)]

plt.figure(figsize = (20, 7))
plt.subplot(121)
plt.imshow(imagenColor)
plt.title('Imagen a Color')
plt.subplot(122)
plt.imshow(imagenGrises, cmap = plt.cm.gray)
plt.title('Imagen en Escala de Grises')
plt.show()

plt.figure(figsize = (20, 7))
plt.subplot(121)
plt.imshow(imagenColorZonaCentral)
plt.title('Imagen a Color (Zona Central)')
plt.subplot(122)
plt.imshow(imagenGrisesZonaCentral, cmap = plt.cm.gray)
plt.title('Imagen en Escala de Grises (Zona Central)')
plt.show()
```

```python
imagenColorCanalRojo = obtenerImagenRGBCanalizada(imagenColor, "R")            
imagenColorCanalVerde = obtenerImagenRGBCanalizada(imagenColor, "G")
imagenColorCanalAzul = obtenerImagenRGBCanalizada(imagenColor, "B")

plt.figure(figsize = (20, 7))
plt.subplot(121)
plt.imshow(imagenColor)
plt.title('Imagen a Color')
plt.subplot(122)
plt.imshow(imagenColorCanalRojo)
plt.title('Imagen a Color (Canal Rojo)')
plt.show()

plt.figure(figsize = (20, 7))
plt.subplot(121)
plt.imshow(imagenColorCanalVerde)
plt.title('Imagen a Color (Canal Verde)')
plt.subplot(122)
plt.imshow(imagenColorCanalAzul)
plt.title('Imagen a Color (Canal Azul)')
plt.show()
```

```python
coordenadasEjeX, coordenadasEjeY, coordenadasEjeZCanalRojo = construirListasCoordenadasDeMatriz(imagenColor[:, :, 0], altoImagen, anchoImagen)
coordenadasEjeX, coordenadasEjeY, coordenadasEjeZCanalVerde = construirListasCoordenadasDeMatriz(imagenColor[:, :, 1], altoImagen, anchoImagen)
coordenadasEjeX, coordenadasEjeY, coordenadasEjeZCanalAzul = construirListasCoordenadasDeMatriz(imagenColor[:, :, 2], altoImagen, anchoImagen)
coordenadasEjeX, coordenadasEjeY, coordenadasEjeZGrises = construirListasCoordenadasDeMatriz(imagenGrises, altoImagen, anchoImagen)

fig = plt.figure(figsize = (23, 14))
ax = fig.add_subplot(111, projection = '3d')
ax.view_init(elev = 20, azim = 345)
ax.scatter(coordenadasEjeX, coordenadasEjeY, coordenadasEjeZCanalRojo, c = 'r', marker = 'o', alpha = 0.125, label = 'Canal Rojo')
ax.scatter(coordenadasEjeX, coordenadasEjeY, coordenadasEjeZCanalVerde, c = 'g', marker = 'o', alpha = 0.125, label = 'Canal Verde')
ax.scatter(coordenadasEjeX, coordenadasEjeY, coordenadasEjeZCanalAzul, c = 'b', marker = 'o', alpha = 0.125, label = 'Canal Azul')
ax.scatter(coordenadasEjeX, coordenadasEjeY, coordenadasEjeZGrises, c = 'k', marker = 'o', alpha = 0.125, label = 'Escala de Grises')
ax.set_xlabel('Posici√≥n de Fila')
ax.set_ylabel('Posici√≥n de Columna')
ax.set_zlabel('Valor de la Se√±al')
plt.title("Visualizaci√≥n como Se√±al (de Tiempo Discreto) de la Imagen por Canales")
plt.tight_layout()
plt.show()
```

```python
valoresEjeX = np.arange(0, altoImagen)
valoresEjeY = np.arange(0, anchoImagen)
valoresEjeX, valoresEjeY = np.meshgrid(valoresEjeX, valoresEjeY)
valoresEjeZCanalRojo = (imagenColor[:, :, 0]).transpose()
valoresEjeZCanalVerde = (imagenColor[:, :, 1]).transpose()
valoresEjeZCanalAzul = (imagenColor[:, :, 2]).transpose()
valoresEjeZGrises = imagenGrises.transpose()

fig = plt.figure(figsize = (23, 14))
ax = fig.add_subplot(111, projection = '3d')
ax.view_init(elev = 20, azim = 345)
ax.plot_surface(valoresEjeX, valoresEjeY, valoresEjeZCanalRojo, color = 'r', alpha = 0.175)
ax.plot_surface(valoresEjeX, valoresEjeY, valoresEjeZCanalVerde, color = 'g', alpha = 0.175)
ax.plot_surface(valoresEjeX, valoresEjeY, valoresEjeZCanalAzul, color = 'b', alpha = 0.175)
ax.plot_surface(valoresEjeX, valoresEjeY, valoresEjeZGrises, color = 'k', alpha = 0.175)
ax.set_xlabel('Posici√≥n de Fila')
ax.set_ylabel('Posici√≥n de Columna')
ax.set_zlabel('Valor de la Se√±al')
plt.title("Visualizaci√≥n como Se√±al (de Tiempo Continuo) de la Imagen por Canales")
plt.tight_layout()
plt.show()
```
