```python
import cv2
from IPython.display import Image
import numpy as np
import matplotlib.pyplot as plt

rutaImagen = 'Figures/Flor (Color).jpg'
anchoVentanaCentral = 25
altoVentanaCentral = 25

imagenColor = cv2.imread(rutaImagen, -1)
imagenColor = cv2.cvtColor(imagenColor, cv2.COLOR_BGR2RGB)
imagenGrises = cv2.imread(rutaImagen, -1)
imagenGrises = cv2.cvtColor(imagenGrises, cv2.COLOR_BGR2GRAY)

altoImagen, anchoImagen = imagenColor.shape[:2]
print("La imagen {} tiene tama√±o {}x{}".format(rutaImagen, anchoImagen, altoImagen))

imagenColorZonaCentral = imagenColor[int((altoImagen - altoVentanaCentral) / 2) : int((altoImagen + altoVentanaCentral) / 2), int((anchoImagen - anchoVentanaCentral) / 2) : int((anchoImagen + anchoVentanaCentral) / 2)]
imagenGrisesZonaCentral = imagenGrises[int((altoImagen - altoVentanaCentral) / 2) : int((altoImagen + altoVentanaCentral) / 2), int((anchoImagen - anchoVentanaCentral) / 2) : int((anchoImagen + anchoVentanaCentral) / 2)]

plt.figure(figsize = (20, 8))
plt.subplot(121)
plt.imshow(imagenColor)
plt.title('Imagen a Color')
plt.subplot(122)
plt.imshow(imagenGrises, cmap = plt.cm.gray)
plt.title('Imagen en Escala de Grises')
plt.show()

plt.figure(figsize = (20, 8))
plt.subplot(121)
plt.imshow(imagenColorZonaCentral)
plt.title('Imagen a Color (Zona Central)')
plt.subplot(122)
plt.imshow(imagenGrisesZonaCentral, cmap = plt.cm.gray)
plt.title('Imagen en Escala de Grises (Zona Central)')
plt.show()


"""
ZonaCentral = img[(N/2)-5:(N/2)+5, (M/2)-5:(M/2)+5]
#print Zona Central
print ("\nValor Pixeles Zona Central: ")
print (ZonaCentral) 
"""
```
