<h1 style="font-size:200%;">Imágenes bitmap.</h1>

<ul style="font-size:170%; line-height : 35px;">
  <li>Es una <span style="color:blue">función $f(x,y)$</span> donde $x$ y $y$ son las coordenas espaciales, y el valor de $f$ en cualquier punto $(x,y)$ es proporcional al brillo o al valor de gris de la imagen en ese punto.</li>
  <li>Pueden pensarse como construidas mediante una gran cantidad de cuadraditos, llamados <span style="color:blue">píxeles</span>, cada uno de ellos rellenos de un color uniforme.</li>
  <li>La calidad de la imagen dependerá de la cantidad de píxeles utilizados para representarla.</li>
  <li>Es el tipo de imagen utilizado para la fotografía y el cine, y el que utilizaremos en este curso.</li>
</ul>

  <li>Las imágenes bitmap no permiten el cambio de escala.</li>
  <li>Al aumentar la escala se produce un efecto conocido como <span style="color:blue">pixelado</span>.</li>
  <li>Dicho efecto se hace más evidente en las líneas curvas y en las zonas en las que hay cambios bruscos de luminosidad.</li>
</ul>

<h1 style="font-size:200%;">Formatos de las imágenes$^1$.</h1>

<h1 style="font-size:200%;"> <span style="color:blue">BMP (Bitmap = Mapa de bits).</span> </h1>

<ul style="font-size:170%; line-height : 35px;">
  <li>Ha sido muy utilizado porque fue desarrollado para aplicaciones Windows.</li>
  <li>La imagen se forma mediante una matriz de píxeles.</li>
  <li>El formato BMP (<span style="color:blue">no sufre pérdidas de calidad</span>) y por tanto resulta adecuado para guardar imágenes que se desean manipular posteriormente.</li>
  <li>Ventaja: Guarda gran cantidad de información de la imagen.</li>
  <li>Inconveniente: El archivo tiene un tamaño muy grande.</li>
</ul>

<h1 style="font-size:300%; background-color:powderblue; color:red; text-align:center; line-height : 60px;">
Imagen digital bi-dimensional </h1>

<h1 style="font-size:200%;"> <span style="color:blue"> GIF (Graphics Interchange Format = Formato de Intercambio Gráfico).</span> </h1>

<ul style="font-size:170%; line-height : 35px;">
  <li>Ha sido diseñado específicamente para comprimir imágenes digitales. </li>
  <li>Reduce la paleta de colores a 256 colores como máximo (profundidad de color de 8 bits). </li>
  <li>Ventaja: Es un formato (<span style="color:blue">idóneo para publicar dibujos en la web</span>). </li>
  <li>Inconveniente: No es recomendable para fotografías de cierta calidad ni originales ya que el color real o verdadero utiliza una paleta de más de 256 colores.</li>
</ul>
<h1 style="font-size:100%; background-color:rgba(70, 0, 700, 0.2); color:black; text-align:center; line-height : 30px;">

<h1 style="font-size:300%; background-color:powderblue; color:red; text-align:center; line-height : 60px;">
Imagen digital bi-dimensional </h1>

<h1 style="font-size:200%;"> <span style="color:blue"> PNG (Portable Network Graphic = Gráfico portable para la red).</span> </h1>

<ul style="font-size:170%; line-height : 35px;">
  <li>Es un formato de reciente difusión alternativo al GIF. </li>
  <li>Tiene una tasa de compresión superior al formato GIF (+10%). </li>
  <li>Admite la posibilidad de emplear un número de colores superior a los 256 que impone el GIF. </li>
  <li>Debido a su reciente aparición (<span style="color:blue">sólo es soportado en navegadores modernos</span>). </li>
</ul>
<h1 style="font-size:100%; background-color:rgba(70, 0, 700, 0.2); color:black; text-align:center; line-height : 30px;">
Visión de Máquina &#8212; F. Prieto </h1>

<h1 style="font-size:300%; background-color:powderblue; color:red; text-align:center; line-height : 60px;">
Imagen digital bi-dimensional </h1>

<h1 style="font-size:200%;"> <span style="color:blue"> JPG-JPEG (Joint Photographic Experts Group = Grupo de Expertos Fotográficos Unidos).</span> </h1>

<ul style="font-size:170%; line-height : 35px;">
  <li>A diferencia del formato GIF, admite una paleta de hasta 16 millones de colores. </li>
  <li>Es el formato (<span style="color:blue">más común junto con el GIF para publicar imágenes en la web</span>). </li>
   <li>Cada vez que se modifica y guarda un archivo JPEG, se puede perder algo de su calidad si se define cierto factor de compresión. </li>
  <li>En la mayoría de los casos esta pérdida se puede asumir porque permite reducir el tamaño del archivo y su visualización es aceptable. </li>
  <li>Las cámaras digitales suelen almacenar directamente las imágenes en formato JPEG con máxima calidad y sin compresión. </li>
  <li>Ventaja: Es ideal para publicar fotografías en la web siempre y cuando se configuren adecuadamente dimensiones y compresión. </li>
  <li>Inconveniente: Si se define un factor de compresión se pierde calidad. Por este motivo no es recomendable para archivar originales. </li>
</ul>

<h1 style="font-size:200%;"> <span style="color:blue"> TIF-TIFF (Tagged Image File Format = Formato de Archivo de Imagen Etiquetada).</span> </h1>

<ul style="font-size:170%; line-height : 35px;">
  <li> Almacena imágenes de una calidad excelente.</li>
  <li>Utiliza cualquier profundidad de color de 1 a 32 bits. </li>
  <li>Es el <span style="color:blue">formato ideal para editar o imprimir una imagen</span>. </li>
  <li>Ventaja: Es ideal para archivar archivos originales. </li>
  <li>Inconveniente: Produce archivos muy grandes. </li>
</ul>

<h1 style="font-size:200%;"> <span style="color:blue">RAW (Formato “en bruto”)$^1$.</span> </h1>

<ul style="font-size:170%; line-height : 35px;">
  <li>Es el formato que ofrece la <span style="color:blue">mayor calidad fotográfica y suele ser admitido por cámaras de gama media y alta</span> (réflex, y compactas), usadas por fotógrafos aficionados avanzados y profesionales.</li>
  <li>Las cámaras que guardan las fotos en otros formatos (Tiff y JPEG), procesan la imagen captada para dar una interpretación de ella (balance de blanco, niveles de luminosidad, contraste…).</li>
  <li>Un archivo RAW, no sufre ninguna compresión, por lo que mantiene el máximo detalle de la imagen a costa de ocupar mucho espacio (Mbytes).</li>
</ul>

<h1 style="font-size:200%;"> <span style="color:blue">Reflexión difusa:</span> </h1>

<ul style="font-size:170%; line-height : 35px;">
  <li>Nos permite distinguir en la <span style="color:blue">vida ordinaria</span> los objetos por difusión directa de la luz en ellos.</li>
  <li>La luz que incide en un objeto se difunde en él y finalmente nuestros ojos y cerebro elaboran la imagen del objeto.</li>
  <li>La luz difundida no cumple con las leyes de reflexión y el proceso de obtención de la imagen es más complejo</li>
</ul>

<h1 style="font-size:200%;"> <span style="color:blue">Reflexión especular:</span> </h1>

<ul style="font-size:170%; line-height : 35px;">
  <li>En esta reflexión se generan imágenes de un objeto al reflejarse la luz proveniente del objeto en una superficie pulida.</li>
  <li>La luz reflejada en dicha superficie converge en algún punto del espacio donde se formará la imagen del objeto, de acuerdo con las leyes de reflexión. </li>
  <li>Estas imágenes pueden ser reales o virtuales.</li>
</ul>

<h1 style="font-size:200%;"> <span style="color:blue">Medidas de distancia.</span> </h1>

<ul style="font-size:170%; line-height : 35px;">
  Dados los píxeles $p=(a,b)$, $q=(s,t)$ y $z=(u,v)$, se dice que $D$ es una función <span style="color:blue">métrica de distancia</span> si cumple las tres propiedades siguientes: 
</ul>

<ul style="font-size:170%; line-height : 35px;">
  <li>$D(p,q) \ge 0, \ (D(p,q) = 0 \ sii \ p=q) \ \rightarrow$ semidefinida positiva.</li>
  <li>$D(p, q) = D(q, p) \ \rightarrow$ simétrica.</li>
  <li>$D(p, z) \le D(p, q) + D(q, z) \ \rightarrow$ desigualdad triangular. </li>
</ul>

<h1 style="font-size:200%;"> <span style="color:blue">Ecualización del Histograma de una imagen.</span> </h1>


<ul style="font-size:170%; line-height : 35px;">
  <li>Es una transformación que busca obtener, para una imagen, un <span style="color:blue">histograma con una distribución uniforme</span>.</li>
  <li>Es decir, que exista el mismo número de píxeles para cada nivel de gris del histograma.</li>
  <li>Está definida como:</li>
</ul>

<ul style="font-size:170%; line-height : 35px; text-align:center; ">
$h(v) =  \text{round} \left (  \frac{cdf(v) - cdf_{min}}  {(M \times N) - 1} 
                \times (L-1) \right ),$              
</ul>

<ul style="font-size:170%; line-height : 35px;">
donde:

<ul>
<li>$v$ es la intensidad de gris del píxel.</li>
<li>$cdf$ es la función de distribución acumulativa.</li>
<li>$cdf_{min}$ es el mínimo valor, diferente de cero, de la función de distribución acumulativa.</li>
<li>$M \times N$ es el número de píxeles en la imagen.</li>
<li>$L$ es el número de niveles de gris.</li>
</ul>

</ul>


<h1 style="font-size:100%; background-color:rgba(70, 0, 700, 0.2); color:black; text-align:center; line-height : 30px;">
Visión de Máquina &#8212; F. Prieto </h1>

```python
#SE ECUALIZA EL HISTOGRAMA DE LA IMAGEN 
equ = cv2.equalizeHist(img)
hist_equ = cv2.calcHist([equ],[0],None,[256],[0,256])

plt.figure(1)
plt.subplot(221), plt.imshow(img, cmap='gray'), plt.title('Img Original')
plt.subplot(222), plt.imshow(equ, cmap='gray'), plt.title('Img Ecualizada')
plt.subplot(223), plt.hist(img.ravel(),128,[0,256])
plt.subplot(224), plt.hist(equ.ravel(),128,[0,256])

# Adjust the subplot layout, because the logit one may take more space
# than usual, due to labels 
plt.subplots_adjust(top=0.95, bottom=0.05, left=0.10, right=0.95, hspace=0.25, wspace=0.35)

plt.show()
```

<h1 style="font-size:200%;"> <span style="color:blue">Histograma: Brillo de una imagen.</span> </h1>


<ul style="font-size:170%; line-height : 35px;">
  <li>Corresponde al nivel gris promedio de la imagen.</li>
  <li>Está definido como:</li>
</ul>

<ul style="font-size:170%; line-height : 35px; text-align:center; ">
$B=\dfrac{1}{NM} \sum\limits_{i=0}^{N-1} \sum\limits_{j=0}^{M-1} I(i,j),$
</ul>

<ul style="font-size:170%; line-height : 35px;">
donde $M$ y $N$ corresponden al alto y ancho de la imagen.
</ul>


<h1 style="font-size:100%; background-color:rgba(70, 0, 700, 0.2); color:black; text-align:center; line-height : 30px;">
Visión de Máquina &#8212; F. Prieto </h1>

```python
#Lee tamaño de la imagen
width, height = img.shape[:2]
#print('N=', width, 'M=', height)

#Calcula el BRILLO
B1=0; B2=0
for i in range(width):
    for j in range(height):
        B1 = B1 + img[i,j]
        B2 = B2 + equ[i,j];
Br1=B1/(width*height)
Br2=B2/(width*height)
#print "Brillo img. original = %d. Brillo img. ecualizada = %d." %(Br1,Br2)


#VISUALIZACION
plt.figure(1)
plt.subplot(121), plt.imshow(img, cmap='gray'), plt.title('Imagen')
plt.subplot(122), plt.imshow(equ, cmap='gray'), plt.title('Img Ecualizada')

# Adjust the subplot layout, because the logit one may take more space
# than usual, due to labels 
plt.subplots_adjust(top=0.95, bottom=0.05, left=0.10, right=0.95, hspace=0.25, wspace=0.35)

plt.show()

print "             Brillo: %d.                        Brillo: %d." % (Br1, Br2)
```

<h1 style="font-size:200%;"> <span style="color:blue">Histograma: Contraste de una imagen.</span> </h1>


<ul style="font-size:170%; line-height : 35px;">
  <li>Corresponde a la variación del nivel de gris en los píxeles de la imagen.</li>
  <li>Está definido como:</li>
</ul>

<ul style="font-size:170%; line-height : 35px; text-align:center; ">
$C = \sqrt{ \dfrac{1}{NM} \sum\limits_{i=0}^{N-1} \sum\limits_{j=0}^{M-1} [I(i,j) - B]^2},$    </ul>

<ul style="font-size:170%; line-height : 35px;">
donde $M$ y $N$ corresponden al alto y ancho de la imagen y $B$ al brillo de la misma.
</ul>


<h1 style="font-size:100%; background-color:rgba(70, 0, 700, 0.2); color:black; text-align:center; line-height : 30px;">
Visión de Máquina &#8212; F. Prieto </h1>

```python
#Calcula el CONTRASTE
B1=0; B2=0
for i in range(width):
  for j in range(height):
    B1=B1 + pow(img[i,j]-Br1,2)
    B2=B2 + pow(equ[i,j]-Br2,2);
Cont1=pow(B1/(width*height),0.5)
Cont2=pow(B2/(width*height),0.5)
#print "Contraste img. original = %d. Contraste img. ecualizada = %d." %(Cont1,Cont2)

#Visualizacion resulatado 

plt.figure(1)
plt.subplot(221), plt.imshow(img, cmap='gray'), plt.title('Original')
plt.subplot(223), plt.hist(img.ravel(),128,[0,256])

plt.subplot(222), plt.imshow(equ, cmap='gray'), plt.title('Ecualizada')
plt.subplot(224), plt.hist(equ.ravel(),128,[0,256])

# Adjust the subplot layout, because the logit one may take more space
# than usual, due to labels 
plt.subplots_adjust(top=0.95, bottom=0.05, left=0.10, right=0.95, hspace=0.25, wspace=0.35)

plt.show()

print "             Contraste = %d.                    Contraste = %d." %(Cont1,Cont2)
```

<h1 style="font-size:200%;"> <span style="color:blue">Operaciones puntuales.</span> </h1>

<ul style="font-size:170%; line-height : 35px;">
  <li>Son operaciones que se <span style="color:blue">aplican a nivel de píxel</span>.</li>
  <li>El resultado en un píxel dado depende únicamente de su valor de intensidad.</li>
  <li>Ejemplos:</li>
  <ul>
      <li>Suma, Resta, Multiplicación, División.</li>
      <li>Máximo, Mínimo.</li>
      <li>Umbralización.</li>

  </ul>  
  Pueden ser con una sola imagen o entre varias imágenes.
</ul>



<h1 style="font-size:100%; background-color:rgba(70, 0, 700, 0.2); color:black; text-align:center; line-height : 30px;">

```python
# DIVISION POR UNA CONSTANTE

# SE CARGAN LAS IMAGENES 
img1 = cv2.imread('Fig/ConceptosBasicos/Imagen_bitmap_1.jpg',0)

width, height = img1.shape[:2]
img_division = np.zeros((width,height), np.uint8)

#DIVISION POR UNA CONSTANTE
C=2 #Valor de la constante
IGMax=0
IGmin=1000000

for i in range(width):
    for j in range(height):
        img_division[i,j] = img1[i,j]/C
        if img1[i,j] > IGMax: IGMax = img1[i,j]
        if img_division[i,j] < IGmin: IGmin = img_division[i,j]

#        print (img1[i,j],img_division[i,j])
#print "Vmax= %d,   Vmin=%d." %(IGMax,IGmin)


#SE CALCULA EL HISTOGRAMA CON LA SIGUIENTE FUNCION
hist1 = cv2.calcHist([img1],[0],None,[256],[0,256])
hist_division = cv2.calcHist([img_division],[0],None,[256],[0,256])


#Visualizacion resulatado 
plt.figure(1)
plt.subplot(221), plt.imshow(img1, cmap='gray', clim=(IGmin, IGMax)), plt.title('Original')
plt.subplot(222), plt.imshow(img_division, cmap='gray', clim=(IGmin, IGMax)), plt.title('Division')

plt.subplot(223), plt.hist(img1.ravel(),256,[0,256])
plt.subplot(224), plt.hist(img_division.ravel(),256,[0,256])


# Adjust the subplot layout, because the logit one may take more space
# than usual, due to labels 
plt.subplots_adjust(top=0.95, bottom=0.05, left=0.10, right=0.95, hspace=0.25, wspace=0.35)

plt.show()

#cv2.imshow('Original',img1)
#cv2.imshow('Division',img_division)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```

<h1 style="font-size:200%;"> <span style="color:blue">Interpolación.</span> </h1>

<ul style="font-size:170%; line-height : 35px;">
<li>La interpolación es un <span style="color:blue">proceso de estimación de valores intermedios de una función continua a partir de muestras discretas</span>.</li>
<li>Es usada intensivamente en el procesamiento de imágenes <b>cuando se cambia el tamaño</b> original de una imagen (ampliación o reducción), o cuando se remapea (distorsiona) la imagen de una malla a otra.</li>
<li>El cambio de tamaño es necesario <b>cuando se incrementa o decrementa el número total de píxeles.</b></li>
<li>El remapeo puede ocurrir en una gran variadad de situaciones: corrección de distorsiones de lentes, cambio de perspectiva, <b>rotación de una imagen</b>, etc.</li>
<ul>
<li>La rotación, alrededor del origen, por un ángulo $\theta$, está dado por:</li>
